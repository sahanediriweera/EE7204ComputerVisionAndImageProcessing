```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
```

```python
img = data.camera()
```

```python
def change_intensity(number_of_intensity_levels,the_image):
  the_entire_imageset = []
  image = the_image
  plt.gray()
  f,arr = plt.subplots(1,number_of_intensity_levels+1,figsize = (20,20))
  arr[0].imshow(the_image)
  for x in range(1,number_of_intensity_levels):
    new_image = np.round((image*(2**(x))/255)).astype(int)
    arr[x].imshow(new_image)
```

```python
change_intensity(4,img)
```

```python
def simple_spacial(img):
  f,arr = plt.subplots(1,4,figsize=(20,20))
  arr[0].imshow(img)
  kernals = [3,10,20]
  for x in range(3):
    kernel = np.ones((kernals[x],kernals[x]),np.float32)/(kernals[x]*kernals[x])
    new_img = cv2.filter2D(img,-1,kernel)
    arr[x+1].imshow(new_img)
  plt.show()
```

```python
simple_spacial(data.astronaut())
```

```python
import scipy.misc
from scipy import ndimage
img = data.lily()
f,arr = plt.subplots(1,2,figsize=(10,10))
rotated_img_45 = ndimage.rotate(img, 45)
arr[0].imshow(rotated_img_45, cmap=plt.cm.gray)
rotated_img_90 = ndimage.rotate(img, 90)
arr[1].imshow(rotated_img_90, cmap=plt.cm.gray)
plt.axis('off')

plt.show()
```

```python
def spatial_avg(img):
  f,arr = plt.subplots(1,4,figsize=(20,20))
  kernals = [3,5,7]
  arr[0].imshow(img)
  for x in range(3):
    new_img = cv2.blur(img, (kernals[x],kernals[x]))
    arr[x+1].imshow(new_img)
  plt.show()
```

```python
spatial_avg(data.rocket())
```
